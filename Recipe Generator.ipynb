```python
#removing the leading two character "c(" and ending ")" from columns

columns_to_clean = ['Images', 'Keywords', 'RecipeIngredientQuantities', 'RecipeIngredientParts', 'RecipeInstructions']

#creating a function to clean values in the specified columns
def clean_values(value):
    if value.startswith('c(') and value.endswith(')'): #checking if the value starts with 'c(' and ends with ')'
        cleaned_value = value[2:-1].strip() #removing leading 'c', '(' and ')' and strip leading/trailing whitespace
        return cleaned_value
    else:
        return value

#Applying the cleaning function to the specified columns
for column in columns_to_clean:
    main_recipeDF[column] = main_recipeDF[column].apply(clean_values)
```

```python
#creating a function that removes the leading two characters if and only if they are "PT"
def remove_pt_prefix(value):
    if isinstance(value, str) and value.startswith("PT"):
        return value[2:]  #removes first two characters
    else:
        return value
```

```python
#applying the remove_pt_prefix function to columns CookTime, PrepTime, TotalTime
main_recipeDF['CookTime'] = main_recipeDF['CookTime'].apply(remove_pt_prefix)
main_recipeDF['PrepTime'] = main_recipeDF['PrepTime'].apply(remove_pt_prefix)
main_recipeDF['TotalTime'] = main_recipeDF['TotalTime'].apply(remove_pt_prefix)
```

```python
#creating a function to reformat the time values
def adjust_time_format(value):
    if pd.isna(value) or value == "NA":
        return "Not Available"
    
    value_str = str(value)
    if len(value_str) > 3:
        if value_str[1] == 'H':
            return value_str[:2] + " " + value_str[2:]
        else:
            return value_str[:3] + " " + value_str[3:]
    else:
        return value_str
```

```python
#applying function to time columns
main_recipeDF['CookTime'] = main_recipeDF['CookTime'].apply(adjust_time_format)
main_recipeDF['PrepTime'] = main_recipeDF['PrepTime'].apply(adjust_time_format)
main_recipeDF['TotalTime'] = main_recipeDF['TotalTime'].apply(adjust_time_format)

#double checking if function worked probably
main_recipeDF
```

# Creating Recipe Generators Based on Certain Criteria

# Keyword Recipe Generator

```python
#creating the random recipe generator based on keywords
def search_recipes_by_keyword(keyword, start_index):
    # Finds recipes containing the keyword in the 'Keywords' column
    matching_recipes = main_recipeDF[main_recipeDF['Keywords'].str.contains(keyword, case=False)]
    
    if not matching_recipes.empty: 
        matching_recipes_count = len(matching_recipes)
        if start_index >= matching_recipes_count:
            print("No more recipes found matching the keyword.")
            return
        
        end_index = min(start_index + 10, matching_recipes_count)
        print(f"Displaying recipes {start_index + 1} to {end_index} out of {matching_recipes_count}:")
        displayed_recipes = matching_recipes.iloc[start_index:end_index]
        
        for index, row in displayed_recipes.iterrows():
            print(f"Recipe Name: {row['Name']}")
            print(f"Description: {row['Description']}")
            print(f"Cook Time: {row['CookTime']}")
            print(f"Prep Time: {row['PrepTime']}")
            print(f"Total Time: {row['TotalTime']}")
            print("---------------------------")
    else:
        print("No recipes found matching the keyword.")

# Interactive loop to allow users to input keywords and search for recipes
start_index = 0
while True:
    user_input = input("Enter a keyword to search for recipes (type 'refresh' to see more, type 'exit' to quit): ").lower()
    if user_input == 'Exit':
        print("Exiting the program.")
        break
    elif user_input == 'refresh':
        start_index += 10
    else:
        search_recipes_by_keyword(user_input, start_index)
```

```python
#function that allows the user to input the desired number of servings from recipe

def find_recipe_by_servings(num_servings, start_index):
    # Find recipes with the specified number of servings
    matching_recipes = main_recipeDF[main_recipeDF['RecipeServings'] == num_servings]
    
    # If there are matching recipes, display 10 recipes starting from start_index
    if not matching_recipes.empty:
        matching_recipes_count = len(matching_recipes)
        if start_index >= matching_recipes_count:
            print("No more recipes found with the specified number of servings.")
            return
        
        end_index = min(start_index + 10, matching_recipes_count)
        print(f"Displaying recipes {start_index + 1} to {end_index} out of {matching_recipes_count}:")
        displayed_recipes = matching_recipes.iloc[start_index:end_index]
        
        for index, row in displayed_recipes.iterrows():
            print(f"Name: {row['Name']}")
            print(f"Description: {row['Description']}")
            print(f"Cook Time: {row['CookTime']}")
            print(f"Prep Time: {row['PrepTime']}")
            print(f"Total Time: {row['TotalTime']}")
            print("---------------------------")
    else:
        print("No recipe found with the specified number of servings.")

#Interactive loop to allow users to input number of servings and search for recipes
start_index = 0
while True:
    user_input = input("Enter the number of servings you desire (type 'refresh' to see more, type 'exit' to quit): ")
    if user_input.lower() == 'exit':
        print("Exiting the program.")
        break
    elif user_input.lower() == 'refresh':
        start_index += 10
    else:
        try:
            num_servings = int(user_input)
            find_recipe_by_servings(num_servings, start_index)
        except ValueError:
            print("Please enter a valid number of servings.")
```

# Creating a function to categorize time values into categories
## applied function to TotalTime column and created a new column from results

```python
#function to categorize time values
def categorize_time(time_value):
    if pd.isna(time_value) or time_value == "NA":
        return "Not Available"
    
    #extracting hours and minutes from the time value
    parts = time_value.split(' ')
    hours, minutes = 0, 0
    for part in parts:
        if 'H' in part:
            hour_part = part.split('H')[0]
            if hour_part.isdigit():
                hours = int(hour_part)
        elif 'M' in part:
            minute_part = part.split('M')[0]
            if minute_part.isdigit():
                minutes = int(minute_part)

    total_minutes = hours * 60 + minutes
    
    #categorize based on total minutes
    if total_minutes <= 15:
        return "15 minutes or less"
    elif total_minutes <= 30:
        return "15-30 minutes"
    elif total_minutes <= 45:
        return "30-45 minutes"
    elif total_minutes <= 60:
        return "45 minutes - 1 hour"
    elif total_minutes <= 120:
        return "1 hour - 2 hours"
    elif total_minutes <= 240:
        return "2 hours - 4 hours"
    elif total_minutes <= 360:
        return "4 hours - 6 hours"
    else:
        return "More than 6 hours"

# Apply the categorization function to the TotalTime column
main_recipeDF['TotalTimeCategory'] = main_recipeDF['TotalTime'].apply(categorize_time)

# Display the updated DataFrame
print(main_recipeDF[['TotalTime', 'TotalTimeCategory']])
```

# Time Category Recipe Generator

```python
# Function to find recipes by time category
def find_recipe_by_time_category(category, offset=0):
    matching_recipes = main_recipeDF[main_recipeDF['TotalTimeCategory'] == category]
    total_matching_recipes = len(matching_recipes)
    if total_matching_recipes > 0:
        print(f"Found {total_matching_recipes} recipes matching the selected time category:")
        for index, row in matching_recipes[offset:offset+10].iterrows():
            print(f"Name: {row['Name']}")
            print(f"Description: {row['Description']}")
            print(f"Cook Time: {row['CookTime']}")
            print(f"Prep Time: {row['PrepTime']}")
            print(f"Total Time: {row['TotalTime']}")
            print("---------------------------")
    else:
        print("No recipes found matching the selected time category.")

# Interactive loop to allow users to select time category and search for recipes
while True:
    print("Select a time category:")
    print("1. 15 minutes or less")
    print("2. 15-30 minutes")
    print("3. 30-45 minutes")
    print("4. 45 minutes - 1 hour")
    print("5. 1 hour - 2 hours")
    print("6. 2 hours - 4 hours")
    print("7. 4 hours - 6 hours")
    print("8. More than 6 hours")
    print("9. Exit")
    
    choice = input("Enter your choice (1-8) or Exit to end program: ")
    
    if choice == 'Exit':
        print("Exiting the program.")
        break
    elif choice in ['1', '2', '3', '4', '5', '6', '7', '8']:
        categories = ["15 minutes or less", "15-30 minutes", "30-45 minutes", 
                      "45 minutes - 1 hour", "1 hour - 2 hours", "2 hours - 4 hours", 
                      "4 hours - 6 hours", "More than 6 hours"]
        selected_category = categories[int(choice) - 1]
        offset = 0
        while True:
            find_recipe_by_time_category(selected_category, offset)
            if len(main_recipeDF[main_recipeDF['TotalTimeCategory'] == selected_category]) <= offset + 10:
                print("No more recipes to display.")
                break
            refresh = input("Enter 'R' to refresh and display 10 new recipes or press Enter to continue: ")
            if refresh.upper() == 'R':
                offset += 10
            else:
                break  # Exit the inner loop if the user doesn't want to refresh
    else:
        print("Invalid choice. Please enter a number between 1 and 9.")
```

# Function that allows user to view all available info for a recipe based on 'Name'; used after other recipe generators

```python
#creating the function "display_recipe_info" for users to use once they have found a recipe they like that will display all
#relevant information for that particular one

def display_recipe_info(recipe_name):
    #finding the recipe with the specified name
    recipe_info = main_recipeDF[main_recipeDF['Name'].str.lower() == recipe_name.lower()]
    
    #if the recipe is found, displays its information as follows
    if not recipe_info.empty:
        print("Recipe Information:")
        for index, row in recipe_info.iterrows():
            print(f"Name: {row['Name']}")
            print(f"Cook Time: {row['CookTime']}")
            print(f"Prep Time: {row['PrepTime']}")
            print(f"Total Time: {row['TotalTime']}")
            print(f"Description: {row['Description']}")
            print(f"Category: {row['RecipeCategory']}")
            print(f"Keywords: {row['Keywords']}")
            print(f"Ingredients (Quantities): {row['RecipeIngredientQuantities']}")
            print(f"Ingredients (Parts): {row['RecipeIngredientParts']}")
            print(f"Rating: {row['AggregatedRating']}")
            print(f"Review Count: {row['ReviewCount']}")
            print(f"Calories: {row['Calories']}")
            print(f"Fat Content: {row['FatContent']}")
            print(f"Saturated Fat Content: {row['SaturatedFatContent']}")
            print(f"Cholesterol Content: {row['CholesterolContent']}")
            print(f"Sodium Content: {row['SodiumContent']}")
            print(f"Carbohydrate Content: {row['CarbohydrateContent']}")
            print(f"Fiber Content: {row['FiberContent']}")
            print(f"Sugar Content: {row['SugarContent']}")
            print(f"Protein Content: {row['ProteinContent']}")
            print(f"Servings: {row['RecipeServings']}")
            print(f"Yield: {row['RecipeYield']}")
            print(f"Instructions: {row['RecipeInstructions']}")
            print("---------------------------")
    else:
        print("Recipe not found.")

# Interactive loop to allow users to input recipe names and display information
while True:
    user_input = input("Enter the name of the recipe (type 'exit' to quit): ")
    if user_input.lower() == 'exit':
        print("Exiting the program.")
        break
    else:
        display_recipe_info(user_input)
```
